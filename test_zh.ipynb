```python
import torch
from face_score import FACEScorer
```

```python
face_scorer = FACEScorer(model_path='/Users/xy/models/gpt2-xl',
                         device='mps')
```

```python
s1 = '请给我杯水'

nll1 = face_scorer.texts_to_nll([s1])
print(nll1)
print('nll1 length:', len(nll1[0]))
```

```python
ids1 = face_scorer.tokenizer.encode(s1)
print('ids1 length:', len(ids1))
print(ids1)

s1_ch1 = face_scorer.tokenizer.decode(ids1[0:2])
s1_ch2 = face_scorer.tokenizer.decode(ids1[2:5])
s1_ch3 = face_scorer.tokenizer.decode(ids1[5:7])
s1_ch4 = face_scorer.tokenizer.decode(ids1[7:9])
s1_ch5 = face_scorer.tokenizer.decode(ids1[9:11])

print(s1_ch1, s1_ch2, s1_ch3, s1_ch4, s1_ch5)
```

```python
s1_ch1_nll = torch.sum(nll1[0][0:1])
s1_ch2_nll = torch.sum(nll1[0][1:4])
s1_ch3_nll = torch.sum(nll1[0][4:6])
s1_ch4_nll = torch.sum(nll1[0][6:8])
s1_ch5_nll = torch.sum(nll1[0][8:10])

print(f'{s1_ch1}     {s1_ch2}     {s1_ch3}      {s1_ch4}      {s1_ch5}')
print(f'{s1_ch1_nll.item():.4f} {s1_ch2_nll.item():.4f} {s1_ch3_nll.item():.4f} {s1_ch4_nll.item():.4f} {s1_ch5_nll.item():.4f}')
```
